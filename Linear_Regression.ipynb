```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.randint(5,35,50).reshape(-1,1)
y = 0.5*X + 3 + np.random.randn(50).reshape(-1,1)
```

```python
plt.scatter(X,y)
plt.title('Data Sample')
plt.xlabel('Independent Values')
plt.ylabel('Dependent Values')
plt.show()
```

```python
lr = LinearRegression() # take regression model to a variable
```

```python
lr.fit(X,y) # train the model
```

```python
y_pred = lr.predict(X)
print(y_pred)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred)
plt.legend(['Actual','Predicted'])
plt.title('Data and Prediction Sample')
plt.xlabel('Independent Values')
plt.ylabel('Dependent Values')
plt.show()
```

```python
lr.coef_ # slope
```

```python
lr.intercept_ # intercept
```
